#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
from math import e
fig, ax = plt.subplots(ncols=2, nrows=1)
ax[0].plot(x,x**2,label="quadratic")
ax[1].plot(x, e**x,label="log")
ax[1].set_xscale("log")
ax[1].set_yscale("log")
```

#### As a bonus challenge, try to add a legend to the plot.

```python
from math import e
fig, ax = plt.subplots(ncols=2, nrows=1)
ax[0].plot(x,x**2,label="quadratic")
ax[1].plot(x, e**x,label="log")
ax[1].set_xscale("log")
ax[1].set_yscale("log")
ax[0].legend()
ax[1].legend()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
fitbit_per_month=fitbit.groupby(by="Months").median().sort_values(by="Months_encoded")
plt.figure(figsize=[20,4])
plt.plot(fitbit_per_month.index,fitbit_per_month["Steps"])
plt.savefig("figures/fitbit_per_month.jpeg")
plt.show()
```

```python
# MEDIAN STEPS BY WORK_OR_WEEKEND
fitbit_per_work_or_weekend=fitbit.groupby(by="Work_or_Weekend").median()
plt.figure(figsize=[2,4])
plt.bar(fitbit_per_work_or_weekend.index,fitbit_per_work_or_weekend["Steps"], width=0.5)
plt.xticks(ticks=[0,1], labels=["Weekend","Work"])
plt.savefig("figures/fitbit_per_work_or_weekend.jpeg")
plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
scatters = ["Minutes Lightly Active", "Minutes Very Active", "Minutes Sedentary"]
for i in scatters:
    plt.scatter(fitbit[i],fitbit["Steps"])
    plt.xlabel(i)
    plt.ylabel("Steps")
    plt.show()
```

#### Use the most appropriate plot to display the summary statistics of `Age` depending on `Pclass`.

```python
fig, ax = plt.subplots(nrows=len(list(titanic["Pclass"].unique())))
i=0
ax[0].boxplot(titanic.loc[titanic["Pclass"]==2,"Age"])
ax[0].set_xlabel(str(2))
```

```python
fig, ax = plt.subplots(nrows=len(titanic["Pclass"].unique()),figsize=[16,10])
i=0
for pclass in np.sort(titanic["Pclass"].unique()):
    ax[i].boxplot(titanic.loc[titanic["Pclass"]==pclass,"Age"])
    ax[i].set_title(f"Pclass = {pclass}")
    ax[i].set_ylabel("Age")
    i+=1
```
