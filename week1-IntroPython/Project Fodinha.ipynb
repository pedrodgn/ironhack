# Fodinha

## Regras do jogo

- Fodinha é um jogo de cartas jogado com baralho de truco. Na primeira rodada os jogadores recebem duas cartas e fazem as suas pedidas, ou seja, diz o seu resultado provável ao final do turno (no caso de da primeira rodada as possibilidades são de 0 a 2). Assim como no truco é virada uma carta para definir a manilha. 
- O numero total de pedidas entre todos os jogadores deve ser diferente do numero de cartas na rodada. Dessa maneira é certo que algum jogador ira errar a pedida. Cada jogador começa com um determinado numero de vidas. E a cada rodada ele perde vidas caso ele erre o numero de pedidas.
- Por exemplo, se um jogador pedir 2 em uma rodada e não fizer com nenhuma carta, logo ele perderá 2 vidas pois: pedidas-resultado = numero de vidas perdidas
- A cada rodada o numero de cartas aumenta em 1.
- Teremos somente dois jogadores: Usuário x Computador

```python
baralho_value ={"4 de paus":0, "4 de copas":0, "4 de espadas":0,"4 de ouros":0,
          "5 de paus":1, "5 de copas":1, "5 de espadas":1,"5 de ouros":1,
          "6 de paus":2, "6 de copas":2,"6 de espadas":2,"6 de ouros":2,
          "7 de paus":3, "7 de copas":3, "7 de espadas":3,"7 de ouros":3,
          "Q de paus":4, "Q de copas":4, "Q de espadas":4,"Q de ouros":5,
          "J de paus":5, "J de copas":5, "J de espadas":5,"J de ouros":5,
          "K de paus":6, "K de copas":6, "K de espadas":6,"K de ouros":6,
            "A de paus":7, "A de copas":7, "A de espadas":7,"A de ouros":7,
            "2 de paus":8, "2 de copas":8, "2 de espadas":8,"2 de ouros":8,
            "3 de paus":9, "3 de copas":9, "3 de espadas":9,"3 de ouros":9,}
          
```

```python
def mao(computador_mao, usuario_mao, baralho_round):
    """
    Joga a mao entre computador e usuario. 
  
    Cada jogador joga uma carta e a função define o resultado. 
  
    Parametros: 
    computador_mao (list): Lista de cartas na mão do computador
    usuario_mao (int): Lista de cartas na mao do usuario
    baralho_round (dict): Dicionario utilizado para definir os valores das cartas nessa mão
  
    Returns: 
    str: um valor que define o vencedor do confronto
    """
    computador_carta = computador_mao[len(computador_mao)-1]
    computador_mao.pop()
    print(f"Computador jogou: {computador_carta}")
    usuario_carta = input(f"Essa é a sua mao: {usuario_mao}. Deseja jogar qual carta? ")
    while usuario_carta not in usuario_mao:
        usuario_carta=input(f"Essa é a sua mao: {usuario_mao}. Escolha uma carta que esteja na sua mao ")
    usuario_mao.remove(usuario_carta)
    if baralho_round[computador_carta]>baralho_round[usuario_carta]:
        print("Computador venceu essa mao")
        return "C"
    elif baralho_round[computador_carta]<baralho_round[usuario_carta]:
        print("Voce venceu essa mao")
        return "U"
    else:
        print("Essa mao deu empate")
```

```python
def rodada(numero):
    """
    Joga uma rodada entre computador e usuário. 
  
    É definida a manilha, as mãos e as pedidas dos jogadores. 
    Cada jogador joga todas as cartas da sua mão e a função define o numero de vidas perdidas por cada.
  
    Parametro: 
    numero (int): Define o numero de cartas que cada jogador tera no começo da rodada.
    Assim esta definido quantas vezes a função mao() sera rodada.
  
    Returns: 
    list: o primeiro valor da lista é o numero de vidas perdidas pelo usuario, e o segundo valor dessa lista é o numero de 
    vidas perdidas pelo computador nessa rodada
    """
    nojogo = set(random.sample(baralho, numero*2))
    usuario_mao = set(random.sample(nojogo, numero))
    computador_mao = list(nojogo-usuario_mao)
    nojogo = list(nojogo)
    usuario_mao=list(usuario_mao)
    manilha = random.choice(cartas)
    print("---------------------")
    input("NOVA RODADA, pronto? ")
    print(f"A manilha: {manilha}")
    print(f"Sua mao: {usuario_mao}")
    manilha1=baralho_value[f"{manilha} de ouros"]
    i=10
    for naipe in naipes:
        baralho_value[f"{manilha} {naipe}"]=i
        i+=1
    computador_value=sum([baralho_value[carta] for carta in computador_mao])
    computador_pedida = round(np.quantile(range(numero), computador_value/(numero*13)))
    print(f"Pedida do computador: {computador_pedida}")
    usuario_pedida = int(input("Qual a sua pedida? "))
    while computador_pedida + usuario_pedida == numero:
        usuario_pedida = int(input("Voce nao pode pedir esse numero. Faça uma pedida diferente "))
    y=0
    computador_resultado = 0
    usuario_resultado = 0
    while y!= numero:
        resultado = mao(computador_mao, usuario_mao, baralho_value)
        if resultado == "U":
            usuario_resultado+=1
        elif resultado =="C":
            computador_resultado+=1
        y+=1
    for x in naipes:
        baralho_value[f"{manilha} {x}"]=manilha1
    return [usuario_pedida-usuario_resultado,computador_pedida-computador_resultado]
```

```python
def fodinha():
    """
    Joga uma partida de fodinha entre computador e usuário. 
  
    É definido o numero de vidas que cada jogador vai ter nessa partida. E joga quantas rodadas forem necessárias
    até um dos jogadores zerarem suas vidas
  
    Returns: 
    str: O vencedor da partida
    """
    print("BEM VINDO AO FODINHA")
    vidas=int(input("Quer jogar com quantas vidas? "))
    computador_vidas=vidas
    usuario_vidas=vidas
    numero=2
    while computador_vidas>0 and usuario_vidas>0:
        resultado_round=rodada(numero)
        computador_vidas=computador_vidas-abs(resultado_round[1])
        usuario_vidas=usuario_vidas-abs(resultado_round[0])
        print(f"Voce perdeu {abs(resultado_round[0])} vidas. Computador perdeu {abs(resultado_round[1])} vidas.")
        print(f"Voce tem {usuario_vidas} vidas. Computador tem {computador_vidas} vidas")
        numero+=1
    if usuario_vidas<computador_vidas:
        print("VOCE PERDEU!!")
    elif computador_vidas<usuario_vidas:
        print("VOCE GANHOU!!!!")
    else:
        print("EMPATE")
    
```
