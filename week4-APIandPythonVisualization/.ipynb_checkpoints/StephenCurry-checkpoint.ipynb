## Data Cleaning and Manipulation

### NBA Stats per Season

```python
seasons_3 = seasons.loc[:,["Season","3P","3PA","PTS","3P%"]]
seasons_3
```

```python
seasons_3.dtypes
```

```python
#replacing empty values for nan
seasons_3=seasons_3.replace(r'^\s*$', np.nan, regex=True)
```

```python
seasons_3["Season"]=seasons_3.loc[:,"Season"].apply(lambda x: x[0:4])
```

```python
seasons_3["Season"]=seasons_3["Season"].astype(int)
seasons_3["3P"]=seasons_3["3P"].astype(float)
seasons_3["3PA"]=seasons_3["3PA"].astype(float)
seasons_3["PTS"]=seasons_3["PTS"].astype(float)
seasons_3["3P%"]=seasons_3["3P%"].astype(float)
```

```python
seasons_3.dtypes
```

```python
seasons_3 = seasons_3.dropna(subset=["3P"])
```

```python
seasons_3 = seasons_3.rename(columns={"3P%":"3P_pct"})
```

```python
seasons_3=seasons_3.sort_values(by="Season")
```

```python
#Pct_change_3P calculates the increase in 3Ps from an year to another
#3P%_in_PTS calculates the importance of 3Ps in PTS made
seasons_3["Pct_change_3P"]=seasons_3["3P"].pct_change()
seasons_3["3P_pct_in_PTS"]=(seasons_3["3P"]*3)/seasons_3["PTS"]
```

```python
seasons_3
```

## Data Storage

```python
load_dotenv("./.env")
```

```python
server = 'postgresql'
username = 'postgres'
password = os.getenv("password")
ip = 'localhost'
database = 'nba3pts'

# create the engine
engine = db.create_engine(f'{server}://{username}:{password}@{ip}/{database}')

# open the connection
conn = engine.connect()
```

```python
players_season_average_1516.to_sql('players_season_average_1516', con=conn, if_exists='replace', index=False)
```

```python
players.to_sql('players', con=conn, if_exists='replace', index=False)
```

```python
seasons_3.to_sql('seasons_3', con=conn, if_exists='replace', index=False)
```

## Analysis

```python
#Analysis: 1980s: 3P shooting growing, 3 point line was a new thing
#1994-1997: shortened 3 point line. The league gave up on this idea in 1997 (thats why we see a sudden increase and 
#decrease)
#2012 NBA lockout shortened season (maybe small sample, something to study)
#2015 and on: Big and constant increases (Maybe Curry Effect?)
seasons_3=seasons_3.sort_values(by="Season",ascending=False)
```

```python
plt.plot(seasons_3["Season"],seasons_3["3P"])
plt.ylabel('3Ps per team per game')
plt.xlabel('Years')
plt.show()
```

```python
seasons_3
```

```python
#So lets look at the players who dominated the 3 ball shooting in the 2015 and 2016 seasons
#That point above everyone is Stephen Curry
season2015_mask=players_season_average_1516["season"]==2015
plt.scatter(players_season_average_1516.loc[season2015_mask,"games_played"],players_season_average_1516.loc[season2015_mask,"fg3m"])
plt.ylabel('3Ps per player per game')
plt.xlabel("Games played")
plt.title("2015 Season")
plt.show()
```

```python
players_season_average_1516.merge(players, on='player_id', how='left').sort_values(by="fg3m",ascending=False).loc[season2015_mask,:].head(10)
```
