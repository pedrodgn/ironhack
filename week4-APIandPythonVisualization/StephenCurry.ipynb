```python
import pandas as pd
import requests
pd.set_option('max_columns', 32)
```

```python
seasons=pd.read_html("https://www.basketball-reference.com/leagues/NBA_stats_per_game.html")[0]
```

```python
seasons.columns = [i[1] for i in seasons.columns]
```

```python
seasons
```

```python
seasons_3 = seasons.loc[:,["Rk","Season","3P","3PA","PTS","3P%"]]
seasons_3
```

```python
seasons_3 = seasons_3.dropna(subset=["3P"]).dropna(subset=["Rk"])
```

```python
cleaningtable=seasons_3["Rk"]!="Rk"
seasons_3=seasons_3.loc[cleaningtable,:]
```

```python
seasons_3.dtypes
```

```python
seasons_3["Rk"]=seasons_3["Rk"].astype(int)
seasons_3["Season"]=seasons_3["Season"].astype(str)
seasons_3["3P"]=seasons_3["3P"].astype(float)
seasons_3["3PA"]=seasons_3["3PA"].astype(float)
seasons_3["PTS"]=seasons_3["PTS"].astype(float)
seasons_3["3P%"]=seasons_3["3P%"].astype(float)
```
