#### Display the trending Python repositories in GitHub.

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url = 'https://github.com/trending/python?since=daily'
response=requests.get(url)
response
````

````python
# your code here
html = response.content
html
````

````python
soup = BeautifulSoup(html)
repositories_html = soup.find_all("h1",attrs={"class":"h3 lh-condensed"})
repositories = [i.text.strip().replace("/\n\n\n\n      ","/ ") for i in repositories_html]
repositories
````

#### Display all the image links from Walt Disney wikipedia page.

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/Walt_Disney'
responses=requests.get(url)
html = responses.content
html
soup=BeautifulSoup(html)
links_html=soup.find_all("a",attrs={"class":"image"})
links = [re.findall("upload\S+",str(i))[0] for i in links_html]
links
````
