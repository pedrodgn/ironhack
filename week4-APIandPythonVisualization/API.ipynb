```python
def get_answers(fromdatetime, todatetime, pageStack=1):
    """
    Pass both dates parameters in the format:
    YYYYMMMDD
    """
    yearfrom = int(str(fromdatetime)[0:4])
    monthfrom = int(str(fromdatetime)[5:6])
    dayfrom = int(str(fromdatetime)[7:8])
    yearto = int(str(todatetime)[0:4])
    monthto = int(str(todatetime)[5:6])
    dayto = int(str(todatetime)[7:8])
    fromdatetime = datetime.date(yearfrom,monthfrom,dayfrom)
    fromdatetime_unixtime = int(time.mktime(fromdatetime.timetuple()))
    todatetime = datetime.date(yearto,monthto,dayto)
    todatetime_unixtime = int(time.mktime(todatetime.timetuple()))
    answers = pd.DataFrame(SITE.fetch("answers",fromdatetime=fromdatetime_unixtime,todatetime=todatetime_unixtime,page=pageStack)["items"])
    answers["owner"]=answers["owner"].apply(lambda x: x["user_id"])
    answers.rename(columns={"owner":"user_id"},inplace=True)
    return answers
get_answers(20200501,20200531)
```

```python
def get_questions(fromdatetime, todatetime, pageStack=1):
    """
    Pass both dates parameters in the format:
    YYYYMMMDD
    """
    yearfrom = int(str(fromdatetime)[0:4])
    monthfrom = int(str(fromdatetime)[5:6])
    dayfrom = int(str(fromdatetime)[7:8])
    yearto = int(str(todatetime)[0:4])
    monthto = int(str(todatetime)[5:6])
    dayto = int(str(todatetime)[7:8])
    fromdatetime = datetime.date(yearfrom,monthfrom,dayfrom)
    fromdatetime_unixtime = int(time.mktime(fromdatetime.timetuple()))
    todatetime = datetime.date(yearto,monthto,dayto)
    todatetime_unixtime = int(time.mktime(todatetime.timetuple()))
    questions = pd.DataFrame(SITE.fetch("questions",fromdatetime=fromdatetime_unixtime,todatetime=todatetime_unixtime,page=pageStack)["items"])
    questions["owner"]=questions["owner"].apply(lambda x: x["user_id"])
    questions.rename(columns={"owner":"user_id"},inplace=True)
    return questions
get_questions(20200501,20200531)
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
def get_most_voted_question_today(fromdatetime, todatetime, pageStack=1):
    """
    Pass both dates parameters in the format:
    YYYYMMMDD
    """
    yearfrom = int(str(fromdatetime)[0:4])
    monthfrom = int(str(fromdatetime)[5:6])
    dayfrom = int(str(fromdatetime)[7:8])
    yearto = int(str(todatetime)[0:4])
    monthto = int(str(todatetime)[5:6])
    dayto = int(str(todatetime)[7:8])
    fromdatetime = datetime.date(yearfrom,monthfrom,dayfrom)
    fromdatetime_unixtime = int(time.mktime(fromdatetime.timetuple()))
    todatetime = datetime.date(yearto,monthto,dayto)
    todatetime_unixtime = int(time.mktime(todatetime.timetuple()))
    questions = pd.DataFrame(SITE.fetch("questions",fromdatetime=fromdatetime_unixtime,todatetime=todatetime_unixtime,page=pageStack,min=5,tagged="python",sort="votes")["items"])
    return questions.iloc[0,:]
get_most_voted_question_today(20200622,20200622)
```
